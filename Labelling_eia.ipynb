```python
# %load labeler.py
import pandas as pd
import re
import os
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import numpy as np
import logging

logger = logging.Logger('logger')


# FOMC
class Config:
    # dukas_data_dir = 'FOMC/FOMC_XAUUSD'
    # time_csv = 'FOMC/fomc_date_time.csv'
    # output_csv = 'fomc_label_15_strategy1.csv'
    #
    # # for dukas
    # dukas_pattern = r'(_\d{4}.\d{2}.\d{2})' # pattern in dukas name
    # formDate = "%Y.%m.%d"
    # formTime = "%Y.%m.%d %H.%M.%S"
    #
    # # for date time csv
    # dateCol = "date"
    # timeCol = "times"
    # formRe = r"(\d{4}/\d{1,2}/\d{1,2})"
    # formDate1 = "%Y/%m/%d"
    # formTime1 = "%Y/%m/%d %H:%M:%S" # "%Y年%m月%d日 %H:%M:%S"
    #
    # strategy = 'strategy_1'
    #
    # end_time = 15
    # start_time = 0
    # threshold = 2

    dukas_data_dir=r'C:\Users\Dell\Desktop\models\XAUUSD_NF'  #下载的tickdata的文件夹
    time_csv=r'C:\Users\Dell\Desktop\nf_dates.csv' #对应的date和timecsv
    output_csv = 'eia_label_30s.csv' #输出的一个csv，date，label，价差
    # for dukas

    dukas_pattern = r'(_\d{2}.\d{2}.\d{4})'  #tickdata文件的文件名的格式
    formDate = "%d.%m.%Y"
    formTime = "%Y.%m.%d %H.%M.%S"


    # for date time csv
    dateCol = "date"
    timeCol = "time"
    labelCol= "1st_label"
    formRe = r"(\d{4}/\d{1,2}/\d{1,2})"
    formDate1 = "%Y/%m/%d"
    formTime1 = "%Y/%m/%d %H:%M:%S" # "%Y年%m月%d日 %H:%M:%S"

    end_time = 240
    start_time = 0
    strategy = 'strategy_1'

    threshold = 0.02 #设置止盈点

class Labeler:
    def __init__(self):
        self.config = Config

    def match_date_dukas(self, text):
        """Regex matching datetimes from string

        Args:
            text (str): str contains datetime in format _\d{2}.\d{2}.\d{4}

        Returns:
            str: matched str
        """
        pattern = self.config.dukas_pattern
        pattern = re.compile(pattern)
        result = pattern.findall(text)
        if result:
            return result[0][1:]
        else:  # Empty Value, no matched pattern
            return None

    def get_dukas_dates(self, path="Data"):
        """Get datetime obj directly from dukas path, wrap function

        Args:
            path (str, optional): Path of duks tick trade historical files. Defaults to "Data".

        Returns:
            list: sorted dukas trade dates in datetime.datetime format
        """
        file_names = os.listdir(path)
        dukas_dates = []
        for filename in file_names:
            date = self.match_date_dukas(filename)
            assert date, 'please check the dukas_pattern in config'
            dukas_dates.append(datetime.strptime(date, self.config.formDate))

        return dict(zip(dukas_dates, file_names))

    @staticmethod
    def get_date_to_time_(df, dateCol, timeCol, formRe=r"(\d{4}年\d{1,2}月\d{1,2}日)", formDate="%Y年%m月%d日",
                 formTime="%Y年%m月%d日 %H:%M"):
        """ Get time from dataframe

        Args:
            df (_type_): dataframe
            dateCol (_type_): date column name
            timeCol (_type_): time column name
            formRe (regexp, optional): format of date regex sercher. Defaults to r"(\d{4}年\d{1,2}月\d{1,2}日)".
            formDate (str, optional): format of date. Defaults to "%Y年%m月%d日".
            formTime (str, optional): format of time. Defaults to "%Y年%m月%d日 %H:%M".

        Returns:
            dict: dict with datetime, keys as date
        """

        import re
        from datetime import datetime
        temp_dates = []
        temp_val = []
        for i in range(len(df[dateCol])):
            temp_dates.append(datetime.strptime(re.findall(formRe, df[dateCol][i])[0], formDate).date())
            temp_val.append(datetime.strptime(re.findall(formRe, df[dateCol][i])[0] + " " + df[timeCol][i], formTime))

        tempDict = dict(zip(temp_dates, temp_val))
        return tempDict

    def get_date_to_time(self):
        """
        transform time csv to a date: val dict
        """
        df = pd.read_csv(self.config.time_csv, encoding='gbk',engine='python')
        datetime_dict = self.get_date_to_time_(df, dateCol=self.config.dateCol,
                                               timeCol=self.config.timeCol,
                                               formRe=self.config.formRe,
                                               formDate=self.config.formDate1,
                                               formTime=self.config.formTime1)
        return datetime_dict

    def AutoLabelSingle(self, filePath: str, date_time_inves: dict[datetime.date:datetime], start: int, end: int):

        def str_time_pipe(str):
            try:
                temp = re.findall(r"(\d{2}.\d{2}.\d{4}\s\d{2}:\d{2}:\d{1,2}.\d{3})", str)[0]
            except:
                raise ValueError("Unvalid string format input")

            return datetime.strptime(temp, "%d.%m.%Y %H:%M:%S.%f")

        def difference(data):
            abs_value = abs(data[-1] - data[0])
            direction = (data[-1] - data[0]) / abs_value
            return direction, abs_value

        def first_max_difference(data, threshold):
            if not data:
                print("no data!!!!!!!")
                return np.nan, np.nan, True
            for i in range(1, len(data)):
                abs_value = abs(data[i] - data[0])
                direction = (data[i] - data[0]) / abs_value
                if abs_value > threshold:
                    print(f"early stop with threshold {threshold}")
                    return direction, abs_value, True
            abs_value = abs(data[-1] - data[0])
            direction = (data[-1] - data[0]) / abs_value
            print(f"nothing to stop")
            return direction, abs_value, False

        def stock_strategy_1(data, threshold):
            min_price, max_price = float('inf'), float('-inf')
            min_pos, max_pos = 0, 0

            max_end, min_end = 0, 0
            pos_diff, neg_diff = float('-inf'), float('inf')

            start, end, max_diff = 0, 0, 0

            for i in range(len(data)):
                if data[i] < min_price:
                    min_pos = i
                    min_price = data[i]

                if data[i] > max_price:
                    max_pos = i
                    max_price = data[i]

                if data[i] - min_price > pos_diff:
                    pos_diff = data[i] - min_price
                    max_end = i

                if data[i] - max_price < neg_diff:
                    neg_diff = data[i] - max_price
                    min_end = i

                # > threshold ?
                if abs(pos_diff) > abs(neg_diff):
                    max_diff = pos_diff
                    start = min_pos
                    end = max_end
                else:
                    max_diff = neg_diff
                    start = max_pos
                    end = min_end

                if abs(max_diff) > threshold:
                    return max_diff, start, end

            assert abs(max_diff) <= threshold
            if abs(max_diff) < threshold:
                logger.warning(
                    f'From {self.config.start_time} to {self.config.end_time}, threshold {threshold} is too large... Only {max_diff} currently')

                return max_diff, start, end
            

        # Slice Time
        path = filePath
        dateX = datetime.strptime(self.match_date_dukas(path), self.config.formDate).date()
        df = pd.read_csv(path,engine='python')

        critical_time = date_time_inves[dateX]

        temp_data = []
        temp_times = []
        temp_diff= []
        for i in range(len(df["Local time"])):
            if critical_time + timedelta(seconds=end) > str_time_pipe(df["Local time"][i]) > critical_time + timedelta(seconds = start):
                avg= (df["Ask"][i] + df["Bid"][i]) / 2
                temp_data.append(avg)
                temp_times.append(str_time_pipe(df["Local time"][i]).second)
                temp_diff.append((str_time_pipe(df["Local time"][i])-critical_time).total_seconds())
        
        if not temp_data:
            print("Warning! There is no prices!")
            print("." * 100)
        # dire, absv = difference(temp_data)
        # dire, absv, early_flag = first_max_difference(temp_data, threshold=self.config.threshold)

        diff, start, end = stock_strategy_1(data=temp_data, threshold=self.config.threshold)

        try:
            plt.title(f"{critical_time.date()}, Direction: {diff*abs(diff)}, Value: {round(diff, 5)}, {temp_times[start]}, {temp_times[end]}")
            
#             gap_num = 3
#             gap = (end - start) // gap_num
#             gap = gap if gap!=0 else 1
#             assert gap != 0, '.....'

#             k = len(temp_data) // ((end - start) // gap)
#             k = k if k else 1
#             xticks = list(range(0, len(temp_data)+k*2, k))[:gap_num]
#             xlabels = list(range(start, end+gap*2, gap))[:gap_num]
#             assert len(xlabels) == len(xticks), f'impossible =.=, {len(xlabels), len(xticks)}'

#             plt.xlabel(temp_times)

            plt.plot(temp_data)
#             print(len(temp_data))
#             print(len(temp_times))
            x = np.arange(len(temp_data))
#             print(len(x))
#             labels = [temp_times]
            gap = int(0.1*len(temp_data))+1
            plt.xticks(x[::gap],temp_times[::gap])
            plt.scatter(x=[start, end], y=[temp_data[start], temp_data[end]], marker='x', c='r')
            plt.text(x=start, y=temp_data[start], s=str(temp_times[start]), c='b')
            plt.text(x=end, y=temp_data[end], s=str(temp_times[end]), c='b')
            
            temp_times_df= pd.DataFrame(temp_diff, columns=['seconds'])
#             sec_3 = temp_times_df[temp_times_df['seconds'] >= 3].index[0]
#             sec_6 = temp_times_df[temp_times_df['seconds'] >= 6].index[0]
            sec_10 = temp_times_df[temp_times_df['seconds'] >= 10].index[0]
#             sec_15 = temp_times_df[temp_times_df['seconds'] >= 15].index[0]
            sec_30 = temp_times_df[temp_times_df['seconds'] >= 30].index[0]
            sec_60 = temp_times_df[temp_times_df['seconds'] >= 60].index[0]
            sec_120 = temp_times_df[temp_times_df['seconds'] >= 120].index[0]
            sec_180 = temp_times_df[temp_times_df['seconds'] >= 180].index[0]
#             plt.vlines(sec_3, min(temp_data), max(temp_data), color="g", linestyles='dashed')
#             plt.vlines(sec_6, min(temp_data), max(temp_data), color="g", linestyles='dashed')
            plt.vlines(sec_10, min(temp_data), max(temp_data), color="g", linestyles='dashed')
#             plt.vlines(sec_15, min(temp_data), max(temp_data), color="g", linestyles='dashed')
            plt.vlines(sec_30, min(temp_data), max(temp_data), color="g", linestyles='dashed')
            plt.vlines(sec_60, min(temp_data), max(temp_data), color="g", linestyles='dashed')
            plt.vlines(sec_120, min(temp_data), max(temp_data), color="g", linestyles='dashed') 
            plt.vlines(sec_180, min(temp_data), max(temp_data), color="g", linestyles='dashed') 
            print('time',critical_time.date().strftime('%Y%m%d'))

            plt.savefig(f'../nf_plot_4mins/trend_{critical_time.date().strftime("%Y%m%d")}.png',dpi=200)
            plt.show()
            print(f"{critical_time.date()}, Direction: {diff/abs(diff)}, Value: {round(diff, 5)}, {start}, {end}")
                                     
            return diff/abs(diff), diff

                                     
        except:
            print("error")

    def run(self):
        date_to_time = self.get_date_to_time() # date -> date, time
        print('date to time complete')
        date_to_csv = self.get_dukas_dates(self.config.dukas_data_dir) # date -> dukas csv name
        print('date to csv complete')
        result = []
        for date in sorted(date_to_csv, reverse=True):

            try:
                result.append(self.AutoLabelSingle(os.path.join(self.config.dukas_data_dir, date_to_csv[date]), date_to_time, self.config.start_time, self.config.end_time))
            except KeyError:
                print(f'No {date} FOMC')
                result.append(np.nan)

        df = pd.DataFrame(dict(zip([sorted(date_to_csv, reverse=True)[i].date().strftime("%Y/%m/%d") for i in range(len(date_to_csv))], result)))
        df.T.to_csv(self.config.output_csv)

sol = Labeler()
sol.run()
```
